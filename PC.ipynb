# PARTE I

```python
import requests

url = "https://realpython.github.io/fake-jobs/"
respuesta = requests.get(url)

print("Status code:", respuesta.status_code)
if respuesta.status_code == 200:
    print("Conexión exitosa: el servidor respondió correctamente.")
else:
    print("Hubo un prob lema al conectar con el servidor.")

print("URL consultada:", respuesta.url)
print("Content-Type:", respuesta.headers["Content-Type"])
```

# PARTE II


```python
from bs4 import BeautifulSoup
sopa = BeautifulSoup(respuesta.text, "html.parser")

empleos = sopa.find_all("div", class_="card-content")
print(f"Total de empleos encontrados: {len(empleos)}")  # 100 — más de 15

lista_empleos = []

for empleo in empleos:
    titulo = empleo.find("h2", class_="title").get_text(strip=True)
    empresa = empleo.find("h3", class_="company").get_text(strip=True)
    ubicacion = empleo.find("p", class_="location").get_text(strip=True)
    fecha = empleo.find("p", class_="is-small has-text-grey").get_text(strip=True)

    lista_empleos.append({
        "titulo": titulo,
        "empresa": empresa,
        "ubicacion": ubicacion,
        "fecha": fecha,
    })

print(f"Registros procesados: {len(lista_empleos)}")
```

```python
import re
patron_fecha = r'\d{4}-\d{2}-\d{2}'

for empleo in lista_empleos[:5]:  # muestro solo 5 para no llenar la pantalla
    coincide = re.search(patron_fecha, empleo["fecha"])
    print(f"{empleo['titulo']:35} fecha: {empleo['fecha']} → válida: {bool(coincide)}")
```

# Parte III

```python
import pandas as pd

# Crear DataFrame a partir de la lista de empleos
df = pd.DataFrame(lista_empleos)

# Mostrar las primeras filas
print("\nPrimeros registros:")
print(df.head())
```

```python
# Eliminar espacios en blanco
df["titulo"] = df["titulo"].str.strip()
df["empresa"] = df["empresa"].str.strip()
df["ubicacion"] = df["ubicacion"].str.strip()

# Convertir la fecha a formato datetime
df["fecha"] = pd.to_datetime(df["fecha"])

# Crear una nueva columna con el año
df["año"] = df["fecha"].dt.year

print("\nInformación del DataFrame:")
print(df.info())
```

```python
# Estadística 1: Total de ofertas laborales
total_ofertas = len(df)

# Estadística 2: Número de empresas diferentes
empresas_diferentes = df["empresa"].nunique()

# Estadística 3: Las 5 empresas con más ofertas
top_empresas = df["empresa"].value_counts().head(5)

# Mostrar resultados
print("\n========== ESTADÍSTICAS ==========")

print(f"\nTotal de ofertas laborales: {total_ofertas}")

print(f"\nNúmero de empresas diferentes: {empresas_diferentes}")

print("\nTop 5 empresas con más ofertas:")
print(top_empresas)
```

```python
# ==========================
# INTERPRETACIÓN DE RESULTADOS
# ==========================

print("\nInterpretación:")
print("Los resultados muestran que se recuperaron", total_ofertas,
      "ofertas laborales publicadas en la página Fake Jobs.")
print("Además, permiten identificar la cantidad de empresas participantes y cuáles tienen mayor presencia en las vacantes analizadas.")
```
