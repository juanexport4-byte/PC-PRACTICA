# PARTE I

```python
import requests

url = "https://realpython.github.io/fake-jobs/"
respuesta = requests.get(url)

print("Status code:", respuesta.status_code)
if respuesta.status_code == 200:
    print("Conexión exitosa: el servidor respondió correctamente.")
else:
    print("Hubo un prob lema al conectar con el servidor.")

print("URL consultada:", respuesta.url)
print("Content-Type:", respuesta.headers["Content-Type"])
```

# PARTE II


```python
from bs4 import BeautifulSoup
sopa = BeautifulSoup(respuesta.text, "html.parser")

empleos = sopa.find_all("div", class_="card-content")
print(f"Total de empleos encontrados: {len(empleos)}")  # 100 — más de 15

lista_empleos = []

for empleo in empleos:
    titulo = empleo.find("h2", class_="title").get_text(strip=True)
    empresa = empleo.find("h3", class_="company").get_text(strip=True)
    ubicacion = empleo.find("p", class_="location").get_text(strip=True)
    fecha = empleo.find("p", class_="is-small has-text-grey").get_text(strip=True)

    lista_empleos.append({
        "titulo": titulo,
        "empresa": empresa,
        "ubicacion": ubicacion,
        "fecha": fecha,
    })

print(f"Registros procesados: {len(lista_empleos)}")
```

```python
import re
patron_fecha = r'\d{4}-\d{2}-\d{2}'

for empleo in lista_empleos[:5]:  # muestro solo 5 para no llenar la pantalla
    coincide = re.search(patron_fecha, empleo["fecha"])
    print(f"{empleo['titulo']:35} fecha: {empleo['fecha']} → válida: {bool(coincide)}")
```
